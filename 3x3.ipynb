```python
def search_matrix_multiplication(initial_tensor, target_tensor, iteration_limit, file_name, dim):
    tensor = initial_tensor
    file = open("results//" + file_name + ".txt", "w")
    file.write(str(initial_tensor) + "\n")
    file.close()
    for k in ["x","y","z"]:
        for j in range(dim**2):
            term = k + str(j)
            tensor = initial_tensor
            print("----------------------------------------------")
            print(term)
            vectors = []
            for i in range(iteration_limit):
                bqm = towards_user_defined(tensor, target_tensor, dim)
                #v = construct_point_for_energy_evaluation(tensor, x, y, z, dim)
                #print(bqm.energy(v))
                if i == 0:
                    bqm.fix_variable(term, 1)
                sample, energy = solve_bqm_in_leap(bqm, sampler = "Greedy") #LeapHybrid
                if i == 0:
                    sample[term] = 1
                print_solution(sample)
                x, y, z = process_result(sample, dim)
                print(x, y, z)
                if energy in vectors:
                    break
                vectors.append(energy)
                if len(x) + len(y) + len(z) == 0:
                    break


                tensor = (tensor - np.tensordot(x, np.tensordot(y, z, axes=0), axes=0)) % 2
                file = open("results//" + file_name + ".txt", "a")
                file.write(str(x) + " " + str(y) + " " + str(z) + "\n")
                file.write(str(tensor) + "\n")
                file.close()
                if np.array_equal(tensor, target_tensor):
                    print("Algorithm reached the target.")
                    print(term)
                    break
```
